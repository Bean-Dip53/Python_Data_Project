```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

# unts of Job Title

```python
df.pivot_table(index='job_title_short', aggfunc='size')
```

we want to gorup by the job_title-short colume but also want to aggregate it to get the median yearly salary

```python
df.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc='median')
```

what is the median salary for the job title broken down by county

```python
df.pivot_table(values='salary_year_avg', index='job_title_short', columns='job_country', aggfunc='median')
```

formatting

```python
df.pivot_table(values='salary_year_avg', index='job_country', columns='job_title_short', aggfunc='median')
```

top countries with values count

```python
top_countries = df['job_country'].value_counts().head(6).index

top_countries
```

making a pivot table

```python
top_countries = df['job_country'].value_counts().head(6).index

df_job_country_salary = df.pivot_table(
    values='salary_year_avg',
    index='job_country',
    columns='job_title_short',
    aggfunc='median'
)

df_job_country_salary =df_job_country_salary.loc[top_countries]

job_title = ['Data Analyst', 'Data Engineer', 'Data Scientist']

df_job_country_salary = df_job_country_salary[job_title]

df_job_country_salary.plot(kind='bar')
```

clean up

```python
top_countries = df['job_country'].value_counts().head(6).index

df_job_country_salary = df.pivot_table(
    values='salary_year_avg',
    index='job_country',
    columns='job_title_short',
    aggfunc='median'
)

df_job_country_salary =df_job_country_salary.loc[top_countries]

job_title = ['Data Analyst', 'Data Engineer', 'Data Scientist']

df_job_country_salary = df_job_country_salary[job_title]

df_job_country_salary.plot(kind='bar')
plt.ylabel('Median Salary ($USD)')
plt.xlabel(' ')
plt.title('Median Salary by Country and Job Title')
plt.xticks(rotation=45, ha='right')
plt.show
```
