COPY() creates a copy of the dataframe. (do this as to not update the original dataframe all malipulation is done on copy only)

```python
dataset = load_dataset('lukebarousse/data_jobs')
df_original = dataset['train'].to_pandas()
```

```python
df_altered = df_original
```

```python
df_altered.loc[:5,'salary_year_avg']
```

```python
median_salary = df_altered['salary_year_avg'].median()

median_salary
```

```python
df_altered['salary_year_avg'] = df_altered['salary_year_avg'].fillna(median_salary)

df_altered.loc[:5,'salary_year_avg']
```

```python
print(id(df_original))

print(id(df_altered))
```

```python
id(df_original) == id(df_altered)
```

the orignial dataframe was updated the same as the altered because they have the same dataframe id
