# Question
In our project we want to see how in-demand skills are trending for data jobs. But before we do that let's look at:

* General trend of data jobs in demand. We're going to look at it by month.

```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_US = df[df['job_country'] == 'United States'].copy()
```

```python
df_US
```

adding verbage of month name

```python
df_US['job_posted_month'] = df_US['job_posted_date'].dt.strftime('%B')

df_US
```

Pivot Table

```python
df_US_pivot = df_US.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')

df_US_pivot
```

Cleaning up the table

Sorting by month 

```python
df_US_pivot.reset_index(inplace=True)
df_US_pivot['month_no'] = pd.to_datetime(df_US_pivot['job_posted_month'], format='%B')

df_US_pivot['month_no']
```

extract out only the month

```python
df_US_pivot.reset_index(inplace=True)
df_US_pivot['month_no'] = pd.to_datetime(df_US_pivot['job_posted_month'], format='%B').dt.month

df_US_pivot['month_no']
```

sort values by month number and Drop the month_no column

```python
df_US_pivot.reset_index(inplace=True)
df_US_pivot['month_no'] = pd.to_datetime(df_US_pivot['job_posted_month'], format='%B').dt.month
df_US_pivot.sort_values('month_no', inplace=True)
df_US_pivot.set_index('job_posted_month', inplace=True)
df_US_pivot.drop(columns='month_no', inplace=True)

```

```python
df_US_pivot
```

Plot

```python
df_US_pivot.plot(kind='line')
```

Clenaing for top 3 job with the most counts

```python
top_3 = df_US['job_title_short'].value_counts().head(3)
top_3 = top_3.index.to_list()
```

```python
df_US_pivot[top_3].plot(kind='line')
```

clean up

```python
df_US_pivot[top_3].plot(kind='line')
plt.title('Monthly Job Postings for Top Data Jobs in the US')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.legend()
plt.show()
```
